```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.wait import WebDriverWait
import time
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options


service = Service(executable_path="/content/chromedriver.exe")
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--headless')
driver = webdriver.Chrome(service=Service(), options=chrome_options)

data_dict = {}
counter = 1

for website in websites:
    web = website
    driver.get(web)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    contact_info_elements = soup.find_all(class_='contact-info primary')
    name_and_title_elements = soup.find_all(class_='name_and_title')

    for contact_info in contact_info_elements:
        mail_a = contact_info.find('a')
        if mail_a and contact_info:
            href_mail = mail_a.get('href')
            print(f"Email: {href_mail}")
            data_dict[counter] = {'Name': '', 'Email': href_mail, 'Website': web}
        else:
            print(f"Email: No Mail")
            data_dict[counter] = {'Name': '', 'Email': 'No Mail', 'Website': web}

        counter += 1

    for idx, name in enumerate(name_and_title_elements):
        names.append(name.text)
        data_dict[counter - len(name_and_title_elements) + idx]['Name'] = name.text

driver.quit()
print(data_dict)
```

```python
import requests
from bs4 import BeautifulSoup
url = 'https://med.stanford.edu/neurology/faculty/overview.html?tab=proxy'

response = requests.get(url)
names = []
websites = []

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    elements = soup.find_all(class_='text col-md-8')
    for element in elements:
        anchor_tags = element.find_all('a')
        for anchor_tag in anchor_tags:
            print(anchor_tag.text)
            names.append(anchor_tag.text)
            href_value = anchor_tag.get('href')
            if href_value:
                websites.append(href_value)

else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```
